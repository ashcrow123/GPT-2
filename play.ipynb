```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel
```

```python
model_path="./gpt2"
model_hf=GPT2LMHeadModel.from_pretrained(model_path,local_files_only=True)
sd_hf=model_hf.state_dict()
for k,v in sd_hf.items():
    print(k, v.shape)
```

```python
from transformers import pipeline
model_path="./gpt2"
generator=pipeline('text-generation',model=model_path,tokenizer=model_path,device=0)
output=generator("The meaning of life is",max_length=30,num_return_sequences=5)
for generate_text in output:
    print(generate_text['generated_text'])
    print("------------------------------------------------------------------------------------------------------------------------")
```
